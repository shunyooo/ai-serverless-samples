# SDXL Sample

```python
prompt = "photo of a huge white cat sitting on a cloud in the sky"
negative_prompt = "blurry, blur, text, watermark, render, 3D, NSFW, nude, CGI, monochrome, B&W, cartoon, painting, smooth, plastic, blurry, low-resolution, deep-fried, oversaturated"
```

### Call via Modal SDK

```python
# AppIdと関数名を指定してどこからでも Python の関数のように呼び出せる
import modal
f = modal.Cls.lookup("stable-diffusion-xl", "Model").inference
image = f.remote(prompt=prompt, negative_prompt=negative_prompt)
# Serialize もよしなにやってくれるのはグッド
image.resize((512, 512))
```

## Call via Web API

```python
import requests
import io
from PIL import Image as PILImage
# Web Endpoint としてデプロイすれば、普通の Web API としても使える
url = "https://clay--stable-diffusion-xl-model-web-inference.modal.run"
res = requests.get(url, params={"prompt": prompt, "negative_prompt": negative_prompt})
image = PILImage.open(io.BytesIO(res.content))
image.resize((512, 512))
```
